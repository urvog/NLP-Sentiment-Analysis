```python
import pandas as pd
import numpy as np
```

```python
#Importando todos los datasets indidivuales por sentimiento
df_pred_anger = pd.read_csv('predictions/pred_anger.csv')
df_pred_anticipation = pd.read_csv('predictions/pred_anticipation.csv')
df_pred_disgust = pd.read_csv('predictions/pred_disgust.csv')
df_pred_fear = pd.read_csv('predictions/pred_fear.csv')
df_pred_joy = pd.read_csv('predictions/pred_joy.csv')
df_pred_love = pd.read_csv('predictions/pred_love.csv')
df_pred_optimism = pd.read_csv('predictions/pred_optimism.csv')
df_pred_pessimism = pd.read_csv('predictions/pred_pessimism.csv')
df_pred_sadness = pd.read_csv('predictions/pred_sadness.csv')
df_pred_surprise = pd.read_csv('predictions/pred_surprise.csv')
df_pred_trust = pd.read_csv('predictions/pred_trust.csv')
```

# Juntando todos los dataset

```python
df_all = df_pred_anger
df_all['anticipation'] = df_pred_anticipation.anticipation
df_all['disgust'] = df_pred_disgust.disgust
df_all['fear'] = df_pred_fear.fear
df_all['joy'] = df_pred_joy.joy
df_all['love'] = df_pred_love.love
df_all['optimism'] = df_pred_optimism.optimism
df_all['pessimism'] = df_pred_pessimism.pessimism
df_all['sadness'] = df_pred_sadness.sadness
df_all['surprise'] = df_pred_surprise.surprise
df_all['trust'] = df_pred_trust.trust
```

```python
df_all.head()
```

```python
#Filtramos sentimientos con un valor pivote y cambiamos el valor a boolean
FILTER_SENTIMENT = 0.75
df_all.anger = [True if x>FILTER_SENTIMENT else False for x in df_all.anger]
df_all.anticipation = [True if x>FILTER_SENTIMENT else False for x in df_all.anticipation]
df_all.disgust = [True if x>FILTER_SENTIMENT else False for x in df_all.disgust]
df_all.fear = [True if x>FILTER_SENTIMENT else False for x in df_all.fear]
df_all.joy = [True if x>FILTER_SENTIMENT else False for x in df_all.joy]
df_all.love = [True if x>FILTER_SENTIMENT else False for x in df_all.love]
df_all.optimism = [True if x>FILTER_SENTIMENT else False for x in df_all.optimism]
df_all.pessimism = [True if x>FILTER_SENTIMENT else False for x in df_all.pessimism]
df_all.sadness = [True if x>FILTER_SENTIMENT else False for x in df_all.sadness]
df_all.surprise = [True if x>FILTER_SENTIMENT else False for x in df_all.surprise]
df_all.trust = [True if x>FILTER_SENTIMENT else False for x in df_all.trust]
```

# Guardando los resultados finales

```python
df_all.to_csv('results/soluciones_mauricio_arancibia_patzi.csv', index = False)
```
